# Setting Up Weekly Meal Schedule

```python
# Calculating the total amount of ingridients needed for a week
Monday    = StirFriedBeetTofu
Tuesday   = Improv
Wednesday = TofuTomatoEggDropSoup
Thursday  = Out
Friday    = TofuTomatoEggDropSoup
Saturday  = Pizza
Sunday    = StirFriedBeetTofu

Week = [Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday]
```

# Computation:
### Shopping Cart, Approximative Cost, Update Avaialble Quantities, Predict Remaining Quantities

```python
# Seeting Ingidients equal to Ingridients in the Monday Meal

Ingridients = Monday

# Iterating through the remaining days (tuesdeay, wednesday, ..., sunday) 
# and adding to the Ingrideints dataframe

for day in range(1,7):
    Ingridients = Ingridients.add(Week[day], fill_value=0)

# Creating a shopping Cart

StockDifference = CurrentStock.add(-Ingridients, fill_value=0)

price = 0
ShoppingCart = pd.DataFrame({'Price': [price]})

for ingridient in Ingridients.columns:
    Supply = StockDifference[ingridient]>0
    if Supply.bool() == True:
        # positve supply 
        PredictingStock[ingridient] = CurrentStock[ingridient] - Ingridients[ingridient] #update expected value for next time    
    else:
        # negative supply --> must buy
        Item = pd.DataFrame({str(ingridient) : [1]})
        NumberItems = np.ceil(Ingridients[ingridient][0])
        
        ShoppingCart = ShoppingCart.add(Item*NumberItems, fill_value=0)
        price = price +  NumberItems * ProductPrice[ingridient][0]
        
        CurrentStock[ingridient] = CurrentStock[ingridient] + ShoppingCart[ingridient]
        PredictingStock[ingridient] = CurrentStock[ingridient]  - Ingridients[ingridient] #update expected value for next time
        
ShoppingCart["Price"] = price
```
